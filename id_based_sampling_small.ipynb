```python
folders = ["test", "train", "validation"]
illness = ["CNV", "DME", "DRUSEN", "NORMAL"]

for folder in folders:
    for ill in illness:
        directory = "OCT_sampling/" + folder + "/" + ill + "/*"

        mylist = glob.glob(directory)
        ids = [i.split("-")[1] for i in mylist]
        ids.sort()
        ids = list(set(ids))


        for x in ids: 
            dummydir = "OCT_sampling/" + folder + "/" + ill + "/" + ill + "-" + x + "-" + "*"
            dummyls = glob.glob(dummydir)


            if len(dummyls) == 1 : 
                movefrom = dummyls[0]
                moveto = "OCT_small" + movefrom.split("OCT_sampling")[1]
                os.rename(movefrom, moveto)


            elif len(dummyls) > 1 :     
                random.seed(77)
                randomls = random.sample(dummyls, 2)
                print(randomls)

                for ls in randomls : 
                    moveto = "OCT_small" + ls.split("OCT_sampling")[1]
                    os.rename(ls, moveto)

            else :
                continue
```
